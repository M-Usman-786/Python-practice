```python
ks_1.head()
```

```python
ks_3 = ks_1[['age in days', 'fare']]
ks_3.head()
```

# -Method of naormalization
- Simple feature scaling
    x(new) = x(old)/x(max)
- min max method
- z-score(standard score -3-to-+3)
- Log transformation

```python
#Simple feature scaling
ks_3['fare'] = ks_3["fare"]/ks_3['fare'].max()
ks_3.head()
```

```python
#max method
ks_3['fare'] = (ks_3["fare"]-ks_3['fare'].min()) / ks_3['fare'].max()-ks_3['fare'].min()
ks_3.head()
```

```python
#z-score (standared score)
ks_3['fare'] = (ks_3["fare"]-ks_3['fare'].mean()) / ks_3['fare'].std()
ks_3.head()
```

```python
ks['fare'] = np.log(ks['fare'])
ks.head()
```
